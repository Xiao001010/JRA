```python
import tifffile
import numpy as np

import os

import matplotlib.pyplot as plt

log_path = 'logs/ResVAE-latentdim128/2023-07-31_14-07-03/'

idx = 500

UnNormalize = True

mean = np.array([6076.685883679818, 1350.9691095158794, 5090.145564947434, 5019.978020658786])
std = np.array([5504.395544161098, 1145.6356702621574, 663.3312283427825, 706.0040270727409])

origin = tifffile.imread(os.path.join(log_path, 'original_{}.tif'.format(idx)))
recon = tifffile.imread(os.path.join(log_path, 'reconstruction_{}.tif'.format(idx)))

print(origin.shape)

if UnNormalize:
    for i in range(4):
        origin[i] = origin[i] * std[i] + mean[i]
        recon[i] = recon[i] * std[i] + mean[i]
```

```python
fig = plt.figure(figsize=(20, 10))
plt.subplot(2, 4, 1)
plt.imshow(origin[0], cmap='gray')
plt.colorbar()
plt.subplot(2, 4, 2)
plt.imshow(origin[1], cmap='gray')
plt.colorbar()
plt.subplot(2, 4, 3)
plt.imshow(origin[2], cmap='gray')
plt.colorbar()
plt.subplot(2, 4, 4)
plt.imshow(origin[3], cmap='gray')
plt.colorbar()
plt.subplot(2, 4, 5)
plt.imshow(recon[0], cmap='gray')
plt.colorbar()
plt.subplot(2, 4, 6)
plt.imshow(recon[1], cmap='gray')
plt.colorbar()
plt.subplot(2, 4, 7)
plt.imshow(recon[2], cmap='gray')
plt.colorbar()
plt.subplot(2, 4, 8)
plt.imshow(recon[3], cmap='gray')
plt.colorbar()
plt.show()
```
